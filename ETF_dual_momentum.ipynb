```python
url = 'https://finance.naver.com/api/sise/etfItemList.nhn'
json_data = json.loads(requests.get(url).text)
```

```python
# ETF 종목 코드 list로 변환
code = df_etf['itemcode'].astype('string').to_list()
```

```python
universe = code 
```

```python
# 종목 Universe 구성

universe_list = universe
```

```python
# 백테스트를 위한 시작, 끝 시점 설정 및 3개월마다 리밸런싱
df = pd.DataFrame(data=None)
for code in universe_list:
        df[code] = fdr.DataReader(code,'2020-05-01','2021-08-31')['Close'].resample('3M').last()
        
```

```python
df
```

```python
# NA가 중간에 있는 종목 제거
df = df.dropna(axis=1)
```

```python
# ETF 종목별 3개월 수익률 계산

df_rtn = df/df.shift(1)*100-100
df_rtn
```

```python
df_rtn = df_rtn.dropna()
df_rtn
```

```python
# 각 기간별 최고 수익률을 보인 종목 선정 - 상대 모멘텀

df_rtn['relative']=df_rtn.idxmax(axis=1)
df_rtn
```

```python
# 최고 수익률을 보인 종목이 양의 수익률이면 True 아니면 False - 절대 모멘텀
df_rtn['absolute'] = df_rtn.max(axis=1)>0
df_rtn
```

## 듀얼 모멘텀 전략 적용을 위한 함수

```python
# 듀얼 모멘텀을 위한 함수 

def dual_momentum_simulation(universe_list):
    ## universe 주가 종목 불러오기
    df = pd.DataFrame(data=None)
    for code in universe_list:
        df[code] = fdr.DataReader(code,'2020-05-01','2021-08-31')['Close'].resample('3M').last()
    
    df=df.dropna(axis=1) # NA 있는 ETF 제거
    ## 수익률 DataFrame으로 가공 후, 상대 절대 모멘텀 계산
    df_rtn = df/df.shift(1)*100-100
    df_rtn = df_rtn.dropna()
    df_rtn['relative'] = df_rtn.idxmax(axis=1)
    df_rtn['absolute'] = df_rtn.max(axis=1)>0
    
    # 듀얼 모멘텀 수익률 계산
    rtn = []
    for i in range(df_rtn.shape[0]-1):
        if df_rtn.iloc[i,-1]:
            rtn.append(df_rtn.iloc[i+1][df_rtn.iloc[i,-2]])
        else:
            rtn.append(0)
    rtn.append(0)
    df_rtn['inv_rtn']=rtn
    df_rtn['cum_rtn'] = ((df_rtn['inv_rtn']/100+1).cumprod()-1)*100
    cum = df_rtn['cum_rtn'].iloc[-2]
    
    print(df_rtn[['relative','inv_rtn','cum_rtn']])
    print(f'----> 전략 누적 수익률:{round(cum,2)}%')
    
    return df_rtn['cum_rtn']
```

```python
universe_list = universe
```

```python
rtn = dual_momentum_simulation(universe_list)
```
